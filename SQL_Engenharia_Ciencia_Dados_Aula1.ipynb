## Linguagem de Definição de Dados (DDL)
A **Linguagem de Definição de Dados** (DDL, do inglês _Data Definition Language_) é um subconjunto do SQL utilizado para definir a estrutura dos dados em um banco de dados. As instruções DDL permitem criar, modificar e excluir objetos de banco de dados, como tabelas e índices. A DDL também permite alterar ou remover dados de maneira mais estrutural.

## Análise dos Resultados Obtidos

Agora que já temos alguns dados na tabela clientes, vamos realizar uma análise e visualizar os resultados, buscando insights que podem ser aplicados em um cenário empresarial.

### 1. Distribuição de Limites de Crédito por Faixa Etária

Empresas que trabalham com crédito geralmente querem entender como o limite de crédito está distribuído por diferentes faixas etárias. Isso ajuda a identificar quais grupos etários recebem maiores limites e podem demandar atenção especial em termos de risco ou oportunidades.

```python
%%sql
SELECT idade, limite_credito FROM clientes;
```

#### Análise:

Podemos plotar um gráfico de dispersão que relacione a idade com o limite de crédito:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Executar consulta para obter dados
result = %sql SELECT idade, limite_credito FROM clientes;

# Convertendo o resultado em DataFrame
df = pd.DataFrame(result, columns=['idade', 'limite_credito'])

# Criando o gráfico de dispersão
plt.figure(figsize=(8,5))
plt.scatter(df['idade'], df['limite_credito'], color='blue')
plt.title('Distribuição do Limite de Crédito por Idade')
plt.xlabel('Idade')
plt.ylabel('Limite de Crédito')
plt.grid(True)
plt.show()
```

O gráfico de dispersão acima mostra a relação entre a idade dos clientes e seus limites de crédito. Esta análise pode ser útil para empresas que concedem crédito, pois oferece uma visão clara de como os limites são distribuídos ao longo de diferentes faixas etárias. A partir do gráfico, podemos observar algumas tendências importantes:

#### Limites de Crédito Maiores para Clientes Mais Jovens
Clientes na faixa etária de 25 a 35 anos tendem a possuir limites de crédito mais elevados, com alguns alcançando valores próximos a 25.000 unidades monetárias.

Essa faixa de clientes representa um segmento que, geralmente, tem um maior potencial de consumo, possivelmente por estar em início ou meio de suas carreiras e ser mais propenso a adquirir bens e serviços, o que justificaria a concessão de maiores limites.

#### Declínio Gradual do Limite de Crédito com o Aumento da Idade
Existe uma redução gradativa do limite de crédito para clientes na faixa etária de 40 a 50 anos. A maioria dos clientes nessa faixa apresenta limites de crédito abaixo de 15.000 unidades monetárias, com poucos casos ultrapassando essa marca.

Este comportamento pode estar relacionado a uma política de crédito mais conservadora para pessoas de meia-idade, onde empresas de crédito podem avaliar que o risco de inadimplência ou diminuição da capacidade de pagamento aumenta com a idade.

#### Clientes Mais Velhos com Limites de Crédito Inferiores
A partir dos 50 anos, o gráfico mostra um claro declínio nos limites de crédito, com clientes dessa faixa geralmente recebendo limites menores, que variam entre 5.000 e 10.000 unidades monetárias, e alguns caindo até abaixo de 5.000.

Isso pode refletir uma política de crédito mais restritiva para clientes mais velhos, que podem ser vistos como de maior risco ou com menor capacidade de gerar renda em longo prazo, especialmente se estiverem próximos da aposentadoria ou em transição para ela.

#### Ausência de Correlação Fortemente Linear
O gráfico não mostra uma correlação linear forte entre idade e limite de crédito. Embora exista uma tendência de queda no limite à medida que a idade aumenta, há uma dispersão significativa nas idades mais jovens (25-35 anos) e uma maior concentração de limites menores nas idades mais avançadas.

Essa dispersão sugere que outros fatores, além da idade, influenciam fortemente o limite de crédito concedido, como nível de renda, histórico de crédito, e ocupação.

#### Implicações Empresariais
- Oportunidades de Mercado: O segmento jovem (25-35 anos) é um grupo que está sendo oferecido limites de crédito mais elevados. Isso pode sugerir que há confiança no poder aquisitivo desse grupo, possivelmente por causa da fase de crescimento profissional em que eles se encontram. Empresas de crédito poderiam explorar esse segmento oferecendo produtos financeiros mais personalizados, como cartões de crédito premium ou opções de crédito diferenciadas.

- Gestão de Risco: A redução do limite de crédito em faixas etárias mais avançadas pode estar relacionada a uma abordagem mais cautelosa. No entanto, isso pode também gerar oportunidades para melhorar a retenção de clientes mais velhos, oferecendo limites ajustados às suas necessidades e uma maior oferta de créditos planejados, como linhas de crédito especial para aposentados.

- Política de Crédito: A empresa pode considerar analisar mais variáveis além da idade, como escolaridade, número de dependentes, e renda mensal para ajustar melhor os limites de crédito, de modo a criar uma relação mais precisa entre perfil do cliente e seu limite.

#### Recomendações:

- Refinamento do Modelo de Concessão de Crédito: Implementar modelos preditivos mais sofisticados que incorporem não apenas a idade, mas também fatores socioeconômicos e de comportamento de pagamento para definir limites de crédito mais justos e eficientes.

- Segmentação Personalizada: Desenvolver produtos e serviços financeiros personalizados para diferentes faixas etárias, com base em seu perfil de risco, e focar em melhorar a retenção de clientes nas faixas de idade mais avançada.

### 2. Média de Limites de Crédito por Escolaridade

Outro aspecto importante é analisar se o nível de escolaridade tem impacto no limite de crédito concedido. Podemos calcular a média dos limites de crédito por nível de escolaridade.

```python
%%sql
SELECT escolaridade, AVG(limite_credito) AS media_limite FROM clientes GROUP BY escolaridade;
```

#### Análise:

```python
# Executar a consulta SQL para obter a média de limite de crédito por escolaridade
result = %sql SELECT escolaridade, AVG(limite_credito) AS media_limite FROM clientes GROUP BY escolaridade;

# Convertendo o resultado em DataFrame
df = pd.DataFrame(result, columns=['escolaridade', 'media_limite'])

# Ordenando os valores de 'media_limite' de forma decrescente
df = df.sort_values(by='media_limite', ascending=False)
```

```python
# Criando o gráfico de barras
plt.figure(figsize=(8,5))
df.plot(kind='bar', x='escolaridade', y='media_limite', legend=False, color='green')
plt.title('Média de Limite de Crédito por Escolaridade (Decrescente)')
plt.xlabel('Escolaridade')
plt.ylabel('Média do Limite de Crédito')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

O gráfico acima mostra a **distribuição da média do limite de crédito** com base nos diferentes níveis de escolaridade, ordenada de maneira decrescente. A análise nos permite identificar o impacto que o nível educacional tem na concessão de crédito, destacando algumas tendências relevantes para instituições financeiras.

#### Doutorado com o Maior Limite de Crédito
Clientes com **doutorado** continuam a liderar, com um limite de crédito médio próximo a **20.000 unidades monetárias**. Isso pode refletir a percepção de que indivíduos com doutorado possuem **maior capacidade de geração de renda**, o que torna esses clientes mais atraentes para as empresas de crédito.

Esses clientes provavelmente ocupam cargos elevados ou possuem estabilidade financeira que justifica a confiança em um limite de crédito mais alto.

#### Ensino Superior Incompleto e Superior Completo com Limites Elevados
Tanto clientes com **ensino superior incompleto** quanto os com **superior completo** têm médias de crédito significativas, entre **15.000 e 17.000 unidades monetárias**.

Esses clientes podem estar em ascensão nas suas carreiras, com o potencial de crescimento profissional e salarial, o que torna esses grupos interessantes para produtos financeiros com limites ajustáveis.

#### Mestrado com Limite Moderado
Os clientes com **mestrado** apresentam uma média de crédito mais modesta, na faixa de **10.000 a 12.000 unidades monetárias**. Curiosamente, este valor está abaixo dos níveis de ensino superior incompleto e completo, o que pode indicar que, no caso desse grupo, outros fatores (além da escolaridade) influenciam na definição do limite, como **empregabilidade** ou **nível de endividamento**.
  
#### Ensino Médio com Limites mais Baixos
O grupo com **ensino médio** apresenta uma média de crédito em torno de **8.000 unidades monetárias**, indicando um limite de crédito mais restrito. Esses clientes podem ser considerados **mais arriscados** devido à **menor qualificação formal**, o que pode estar correlacionado com uma **menor estabilidade no mercado de trabalho**.

#### Ensino Fundamental com o Menor Limite de Crédito
Por fim, clientes com **ensino fundamental** apresentam a média de crédito mais baixa, por volta de **5.000 unidades monetárias**. Isso reflete uma postura conservadora das empresas de crédito, dado que esses indivíduos podem ter **menor poder aquisitivo** e estar em ocupações mais vulneráveis.


#### Implicações Empresariais

- Sgmentação por Escolaridade:
Clientes com doutorado e ensino superior (incompleto ou completo) representam uma **classe de menor risco**, o que permite oferecer **produtos financeiros premium** com maiores limites de crédito e melhores condições de pagamento. Já os clientes com ensino fundamental ou médio podem exigir uma abordagem mais **cautelosa**, com **limites de crédito menores** ou produtos que ajudem a melhorar a saúde financeira do cliente.

- Auste nas Políticas de Crédito:
A relação entre escolaridade e limite de crédito está clara neste gráfico. As empresas podem otimizar suas políticas de concessão de crédito ao avaliar não apenas a escolaridade, mas também **renda**, **ocupação**, e **histórico de pagamento**.
O grupo de clientes com mestrado, por exemplo, pode ser explorado mais a fundo, verificando se há outros fatores de risco que justificam a concessão de limites mais baixos em comparação com os grupos de ensino superior completo ou incompleto.

- Oportunidades de Fidelização:
Clientes com doutorado e ensino superior incompleto apresentam uma alta capacidade de crédito e podem ser alvos para **estratégias de retenção** por meio de cartões de crédito de alto valor ou linhas de crédito personalizadas para **expansão profissional** ou **investimentos**.

Clientes com maior escolaridade, especialmente aqueles com doutorado, tendem a receber os maiores limites de crédito, enquanto clientes com ensino fundamental recebem limites menores, indicando uma **avaliação de risco** mais conservadora por parte das instituições financeiras.

### 3. Impacto do Número de Dependentes no Limite de Crédito

Agora vamos investigar se o número de dependentes de um cliente influencia o limite de crédito concedido.

```python
%%sql
SELECT dependentes, AVG(limite_credito) AS media_limite FROM clientes GROUP BY dependentes;
```

#### Análise

```python
# Executar a consulta SQL
result = %sql SELECT dependentes, AVG(limite_credito) AS media_limite FROM clientes GROUP BY dependentes;

# Convertendo o resultado em DataFrame
df = pd.DataFrame(result, columns=['dependentes', 'media_limite'])
```

```python
# Criando um gráfico de barras
plt.figure(figsize=(8,5))
df.plot(kind='bar', x='dependentes', y='media_limite', legend=False, color='orange')
plt.title('Média de Limite de Crédito por Número de Dependentes')
plt.xlabel('Número de Dependentes')
plt.ylabel('Média do Limite de Crédito')
plt.grid(True)
plt.show()
```

O gráfico acima ilustra a relação entre o **número de dependentes** que um cliente possui e a **média de limite de crédito** concedido a esses clientes. A partir da visualização, podemos identificar algumas tendências importantes que ajudam a entender como as empresas de crédito avaliam o número de dependentes no momento de definir os limites de crédito.

#### Clientes Sem Dependentes Recebem os Maiores Limites de Crédito
- Clientes que **não possuem dependentes** são aqueles que recebem, em média, os maiores limites de crédito, próximos a **20.000 unidades monetárias**.
- Isso pode refletir uma **maior capacidade financeira líquida**, visto que esses indivíduos não possuem obrigações familiares adicionais, como alimentação, educação e outras despesas relacionadas a dependentes. Como resultado, as instituições financeiras veem esses clientes como **menos arriscados** e aptos a receber limites de crédito mais elevados.

#### Queda Gradativa no Limite de Crédito Conforme o Número de Dependentes Aumenta
- Conforme o número de dependentes aumenta, o limite de crédito tende a **cair progressivamente**. Por exemplo:
  - Clientes com **1 ou 2 dependentes** ainda possuem limites relativamente altos, em torno de **15.000 a 17.000 unidades monetárias**.
  - A partir de **3 dependentes**, o limite de crédito começa a cair mais acentuadamente, atingindo uma média de **13.000 unidades monetárias**.
  
- Esse padrão é provavelmente resultado da **maior pressão financeira** sobre o orçamento dos clientes com mais dependentes, o que aumenta o risco percebido de inadimplência por parte das instituições financeiras.

#### Limites de Crédito Mais Baixos para Clientes com 4 ou Mais Dependentes
- Clientes com **4 ou mais dependentes** recebem limites de crédito significativamente menores, na faixa de **5.000 a 10.000 unidades monetárias**.
- Esse comportamento pode estar relacionado ao **risco de endividamento elevado**, pois clientes com muitos dependentes geralmente enfrentam despesas consideráveis com a família, o que pode comprometer sua capacidade de cumprir com obrigações financeiras adicionais.

#### Possíveis Justificativas para Limites de Crédito Menores com Muitos Dependentes
- Clientes com um número maior de dependentes geralmente têm **maiores responsabilidades financeiras**, o que pode reduzir sua **renda disponível** para pagar empréstimos ou utilizar grandes limites de crédito de maneira responsável.
- Empresas de crédito podem ver esses clientes como **mais suscetíveis a dificuldades financeiras**, o que justifica a concessão de limites de crédito mais baixos para evitar riscos elevados de inadimplência.

#### Implicações Empresariais

- Política de Crédito Ajustada ao Número de Dependentes:
A análise mostra que há uma tendência clara de redução de limites de crédito à medida que o número de dependentes aumenta. Instituições financeiras podem ajustar suas políticas para fornecer limites de crédito **mais adequados ao perfil financeiro** de clientes com dependentes, considerando não apenas o número de dependentes, mas também **renda** e **outros indicadores econômicos**.
   
- Educação Financeira para Clientes com Muitos Dependentes:
Para clientes com **4 ou mais dependentes**, limites de crédito mais baixos podem ser acompanhados de **programas de educação financeira**, ajudando esses indivíduos a gerenciar melhor suas finanças e suas responsabilidades familiares, e evitando que acumulem dívidas que não possam pagar.

- Oportunidade para Produtos Financeiros Personalizados:
Para clientes com **1 a 2 dependentes**, que ainda apresentam boa capacidade de crédito, podem ser oferecidos **produtos financeiros personalizados**, como cartões de crédito com limites flexíveis ou linhas de crédito para **despesas familiares específicas**, como educação ou saúde.
Já para clientes com muitos dependentes, produtos que incentivem o **uso responsável de crédito** e limites menores podem ser mais apropriados.

O gráfico revela uma clara **correlação inversa** entre o número de dependentes e o limite de crédito concedido. Clientes sem dependentes tendem a receber os maiores limites de crédito, enquanto aqueles com 4 ou mais dependentes recebem limites significativamente menores. 

Para as instituições financeiras, esses dados podem ser usados para melhorar a segmentação de crédito e ajustar as **políticas de risco** de forma mais eficiente, garantindo que os limites concedidos estejam alinhados com a capacidade financeira e as responsabilidades de cada cliente. Além disso, isso abre oportunidades para oferecer **produtos personalizados** que atendam às necessidades financeiras específicas de cada grupo.

Essa análise reforça a importância de considerar o número de dependentes como uma **variável-chave** no momento de definir os limites de crédito, promovendo a **sustentabilidade financeira** tanto para a instituição quanto para o cliente.

A partir dos resultados obtidos ao longo deste estudo introdutório de SQL, fomos capazes de realizar uma análise detalhada e perspicaz dos dados relacionados a clientes, seus limites de crédito e suas características demográficas, como **idade**, **escolaridade**, e **número de dependentes**. Este exercício demonstrou como o SQL, mesmo em seu uso mais básico, já permite extrair informações críticas e insights valiosos que podem orientar a **tomada de decisões empresariais**.

#### 1. **SQL como Ferramenta Poderosa de Análise de Dados**
- Através de **consultas SQL simples**, como **seleção**, **agregação** e **agrupamento**, foi possível realizar análises profundas e criar visualizações claras que revelam **tendências comportamentais** e **correlações** importantes.
- SQL nos permitiu trabalhar diretamente com os dados armazenados, manipulando grandes volumes de informação de maneira eficiente e organizando-os para extrair valor estratégico para o negócio.
  
#### 2. **Análise Detalhada de Limites de Crédito por Faixa Etária**
- Observamos que clientes mais jovens (25 a 35 anos) tendem a receber limites de crédito mais elevados, sugerindo que este grupo pode ser visto como promissor devido ao seu **potencial de crescimento econômico**.
- Em contrapartida, clientes mais velhos, principalmente aqueles acima de 50 anos, apresentaram limites de crédito menores, o que pode refletir uma **avaliação de risco mais conservadora** por parte das instituições de crédito.
  
#### 3. **Impacto da Escolaridade no Limite de Crédito**
- O gráfico mostrou que **clientes com maior escolaridade** (doutorado, ensino superior completo) têm, em média, maiores limites de crédito. Esse comportamento pode estar relacionado à **expectativa de maior estabilidade financeira** e capacidade de pagamento que esses indivíduos trazem.
- Por outro lado, clientes com **ensino fundamental** ou **médio** têm limites significativamente menores, o que reforça a importância da **escolaridade como um indicador de risco** no processo de concessão de crédito.

#### 4. **Influência do Número de Dependentes**
- Descobrimos que, à medida que o **número de dependentes aumenta**, o limite de crédito tende a diminuir. Clientes sem dependentes apresentaram os maiores limites, enquanto aqueles com **4 ou mais dependentes** receberam limites significativamente menores.
- Isso reflete como o **comprometimento financeiro adicional** associado à criação de dependentes afeta a percepção de risco das instituições financeiras, que tendem a adotar uma abordagem mais conservadora para esses grupos.

### Implicações Empresariais

Com base nos insights extraídos das análises feitas com SQL, fica claro que esta linguagem desempenha um papel muito importante para embasar a **tomada de decisões empresariais**, permitindo às organizações:

1. **Segmentar clientes de maneira eficiente**:
   - Empresas podem usar os dados extraídos de análises SQL para desenvolver **estratégias de segmentação** com base em fatores como idade, escolaridade e número de dependentes, criando produtos financeiros mais ajustados às necessidades de cada grupo.

2. **Aprimorar políticas de crédito**:
   - As análises revelam perfis de risco baseados em dados demográficos, o que possibilita ajustes nas **políticas de concessão de crédito**. Por exemplo, clientes mais jovens ou com maior escolaridade podem ser alvos de limites de crédito mais elevados, enquanto clientes com mais dependentes podem exigir limites mais controlados ou ofertas mais conservadoras.

3. **Personalizar produtos financeiros**:
   - SQL permite identificar padrões que apoiam a criação de **produtos financeiros personalizados**. Por exemplo, clientes sem dependentes e com alta escolaridade podem ser oferecidos produtos de crédito premium, enquanto aqueles com dependentes podem se beneficiar de produtos de crédito mais ajustados a suas responsabilidades familiares.

4. **Gerenciamento de riscos mais assertivo**:
   - Ao usar SQL para analisar múltiplas variáveis (como idade, escolaridade e número de dependentes), as empresas podem **mitigar riscos** ao ajustar os limites de crédito com base em **indicadores-chave**. Isso permite equilibrar o crescimento da carteira de clientes com a **redução de inadimplência**.

Este primeiro contato com a linguagem já mostra sua importância e aplicabilidade para áreas como **finanças, crédito, marketing e operações**, possibilitando um gerenciamento **mais eficiente e estratégico** de negócios baseados em dados.
