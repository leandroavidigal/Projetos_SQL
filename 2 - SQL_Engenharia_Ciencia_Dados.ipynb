## 1. Trabalhando com Tabelas no SQL

### 1.1. Tipos de Dados no SQL

Antes de criar tabelas, é importante conhecer os principais tipos de dados disponíveis no SQL:

- **Números Inteiros:**
  - `INT`: Números inteiros com intervalo de -2.147.483.648 a 2.147.483.647 (4 bytes).
  - `BIGINT`: Inteiros maiores (-9.223.372.036.854.775.808 a 9.223.372.036.854.775.807).
  - `SMALLINT`: Menor intervalo (-32.768 a 32.767).
  - `TINYINT`: Apenas valores de 0 a 255.

- **Números Decimais:**
  - `FLOAT` e `DOUBLE`: Valores de ponto flutuante para dados aproximados.
  - `DECIMAL(precision, scale)`: Dados numéricos exatos, onde `precision` é o total de dígitos e `scale` é o número de dígitos após o ponto decimal.

- **Cadeias de Caracteres:**
  - `CHAR(n)`: Texto de comprimento fixo.
  - `VARCHAR(n)`: Texto de comprimento variável até `n` caracteres.
  - `TEXT`: Texto longo.

- **Datas e Horários:**
  - `DATE`: Apenas data (YYYY-MM-DD).
  - `TIMESTAMP`: Data e hora (YYYY-MM-DD HH:MM:SS).

- **Booleanos:**
  - `BOOLEAN`: Verdadeiro (1) ou Falso (0).

### 1.2. Criando Tabelas

No SQL, usamos o comando `CREATE TABLE` para definir a estrutura de uma tabela.

**Sintaxe Geral:**
```sql
CREATE TABLE <nome_da_tabela> (
  <nome_da_coluna_1> <tipo_da_coluna_1>, 
  <nome_da_coluna_2> <tipo_da_coluna_2>, 
  ...
);
```

**Exemplo 1:**
Criando uma tabela de transações.
```sql
CREATE TABLE transacoes (
    id_transacao INT,
    id_cliente INT,
    data_compra DATE,
    valor FLOAT
);
```

**Exemplo 2:**
Criando uma tabela de clientes.
```sql
CREATE TABLE clientes (
    id INT,
    nome VARCHAR(100),
    data_de_nasc DATE,
    cpf BIGINT,
    filhos BOOLEAN,
    cep INT,
    rua VARCHAR(200)
);
```

### 1.3. Inserindo Dados em Tabelas

Para adicionar registros em tabelas, usamos o comando `INSERT INTO`.

**Exemplo:**
```sql
INSERT INTO clientes (id, nome, data_de_nasc, cpf, filhos, cep, rua)
VALUES (1, 'João Silva', '1990-05-15', 12345678901, TRUE, 12345678, 'Rua A');
```

### 1.4. Atualizando Dados

Para alterar dados existentes em uma tabela, usamos o comando `UPDATE`.

**Exemplo:**
```sql
UPDATE clientes
SET rua = 'Rua B'
WHERE id = 1;
```

### 1.5. Excluindo Dados

Para remover registros de uma tabela, usamos o comando `DELETE`.

**Exemplo:**
```sql
DELETE FROM clientes
WHERE id = 1;
```

## 2. Integração com AWS

Ao trabalhar com bancos de dados na AWS, como o Amazon RDS ou Redshift, é importante conhecer ferramentas e serviços que facilitam a conexão e manipulação de dados.

### 2.1. Configurando o Ambiente

1. **Criar uma Instância RDS:**
   - Acesse o console da AWS.
   - Selecione "RDS" e clique em "Create database".
   - Configure os parâmetros (engine, versão, credenciais, etc.).

2. **Configurar a Conexão:**
   - Use ferramentas como DBeaver, SQL Workbench ou qualquer IDE de sua preferência.
   - Certifique-se de liberar o acesso na "Security Group" associado ao RDS.

3. **Instalar e Configurar AWS CLI:**
   - Instale a CLI seguindo a [documentação oficial](https://docs.aws.amazon.com/cli/latest/userguide/install-cliv2.html).
   - Configure com o comando `aws configure`.

### 2.2. Consultando Dados no Redshift

Com o Redshift, você pode usar o `psql` ou outras ferramentas para executar comandos SQL e manipular grandes volumes de dados.

**Exemplo de Consulta:**
```sql
SELECT nome, salario
FROM funcionarios
WHERE salario > 4000
ORDER BY salario DESC;
```

## 3. Exercícios Práticos

Agora que você revisou os conceitos, é hora de prática!

### Exercício 1: Criação e Inserção de Dados

1. Crie uma tabela chamada `produtos` com as seguintes colunas:
   - `id` (INT, chave primária)
   - `nome` (VARCHAR(100))
   - `categoria` (VARCHAR(50))
   - `preco` (DECIMAL(10, 2))

2. Insira os seguintes registros:
   - (1, 'Notebook', 'Eletrônicos', 3500.00)
   - (2, 'Smartphone', 'Eletrônicos', 2000.00)
   - (3, 'Cafeteira', 'Eletrodomésticos', 250.00)

### Solução:
```sql
CREATE TABLE produtos (
    id INT PRIMARY KEY,
    nome VARCHAR(100),
    categoria VARCHAR(50),
    preco DECIMAL(10, 2)
);

INSERT INTO produtos (id, nome, categoria, preco)
VALUES (1, 'Notebook', 'Eletrônicos', 3500.00),
       (2, 'Smartphone', 'Eletrônicos', 2000.00),
       (3, 'Cafeteira', 'Eletrodomésticos', 250.00);
```

---

### Exercício 2: Consultas com Filtros

Usando a tabela `produtos`, responda:

1. Quais produtos custam mais de R$ 1000,00?
2. Liste os produtos da categoria "Eletrônicos".

### Solução:
```sql
-- Produtos com preco > 1000
SELECT *
FROM produtos
WHERE preco > 1000;

-- Produtos da categoria Eletrônicos
SELECT *
FROM produtos
WHERE categoria = 'Eletrônicos';
```
